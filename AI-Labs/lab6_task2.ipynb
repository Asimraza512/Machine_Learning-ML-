```python
label_encoder=LabelEncoder()
train["Class"]=label_encoder.fit_transform(train["Class"])
train.head()
```

```python
#For Train
x=train.drop("Class",axis=1)
y=train["Class"]
y
```

```python
x
```

```python
train_model=DecisionTreeClassifier(random_state=10)
train_model.fit(x,y)
```

```python
#For prediction
train_pred=train_model.predict(x)
train_pred
```

```python
#For Accuracy
accuracy=accuracy_score(y,train_pred)
accuracy
```
